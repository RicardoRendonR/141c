```python
def kruskal_141C (df):
    groups = np.unique(df)
    N = len(groups)
    
    # create sub dataset
    d_list = []
    for i in range(N):
        d = Y[df== groups[i]] 
        d_list = d_list + [d]
        
    # kruskal wallis test depending on the levels of the target variable
    if N == 2:
        stat, p = kruskal(d_list[0], d_list[1])
        
    elif N == 3:
        stat, p = kruskal(d_list[0], d_list[1], d_list[2])
        
    elif N == 4:
        stat, p = kruskal(d_list[0], d_list[1], d_list[2], d_list[3])
        
    elif N == 5:
        stat, p = kruskal(d_list[0], d_list[1], d_list[2], d_list[3], d_list[4])
        
    elif N == 6:
        stat, p = kruskal(d_list[0], d_list[1], d_list[2], d_list[3], d_list[4], d_list[5])
    
    elif N == 7:
        stat, p = kruskal(d_list[0], d_list[1], d_list[2], d_list[3], d_list[4], d_list[5], d_list[6])
        
    elif N == 8:
        stat, p = kruskal(d_list[0], d_list[1], d_list[2], d_list[3], d_list[4], d_list[5], d_list[6], d_list[7])    
    
    elif N == 9:
        stat, p = kruskal(d_list[0], d_list[1], d_list[2], d_list[3], d_list[4], d_list[5], d_list[6], d_list[7], d_list[8])
    
    elif N == 15:
        stat, p = kruskal(d_list[0], d_list[1], d_list[2], d_list[3], d_list[4], d_list[5], d_list[6], d_list[9], d_list[10], d_list[11], d_list[12], d_list[13], d_list[14])
    
    elif N == 16:
        stat, p = kruskal(d_list[0], d_list[1], d_list[2], d_list[3], d_list[4], d_list[5], d_list[6], d_list[9], d_list[10], d_list[11], d_list[12], d_list[13], d_list[14], d_list[15])
    
    elif N == 25:
        stat, p = kruskal(d_list[0], d_list[1], d_list[2], d_list[3], d_list[4], d_list[5], d_list[6], d_list[9], d_list[10], d_list[11], d_list[12], d_list[13], d_list[14], d_list[15], d_list[16], d_list[17], d_list[18], d_list[19], d_list[20], d_list[21], d_list[22], d_list[23], d_list[24])
    
    result = p
    
    return( result)
    
```
